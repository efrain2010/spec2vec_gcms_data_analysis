The diagonal is ignored every time the percentile value is obtained.

## Obtaining the average scores for the 1st and 0.7th percentiles

```python

matrix_filename = "avg_mol_sim_spec2vec_10iterarions_percentile1.npy"
filename = os.path.join(path, matrix_filename)
similarity_matrix_spec2vec1 = np.load(filename)

matrix_filename = "avg_mol_sim_spec2vec_10iterarions_percentile07.npy"
filename = os.path.join(path, matrix_filename)
similarity_matrix_spec2vec07 = np.load(filename)
```
