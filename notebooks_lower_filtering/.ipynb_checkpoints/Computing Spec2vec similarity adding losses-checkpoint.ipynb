```python
from matchms.importing import load_from_msp

spectrums = [s for s in load_from_msp(msp_file)]
print("Number of Spectra:", len(spectrums))
```

```python
import numpy as np

spectrums_sample = spectrums[0:2]

diffs = []
for spectrum in spectrums_sample:    
    mz = spectrum.peaks[0] # first position is mz
    for i, mz1 in enumerate(mz[:-1]): # loop over peaks from 0 to N-1
        for mz2 in mz[i+1:]:
            diffs.append(abs(int(mz1 - mz2)))
```

```python
for s in diffs:
    print(s)
```

```python
from custom_functions.spectra_functions import grouped_by_same_mol
from custom_functions.spectra_functions import grouped_by_diff_mol

spectrums_diff_mol = grouped_by_diff_mol(spectrums)
spectrums_diff_mol = spectrums_diff_mol[0:2]
```

```python
diffs2 = []
for pairs in spectrums_diff_mol:
    peaks_1 = pairs[0].peaks[0]
    peaks_2 = pairs[1].peaks[0]
    random_peak_1 = np.random.choice(peaks_1, 1)[0]
    random_peak_2 = np.random.choice(peaks_2, 1)[0]
    diffs2.append(int(random_peak_1 - random_peak_2))
```

```python
for s in diffs2:
    print(s)
```

```python
def count_shifts(diffs):
    unique_diffs = []
    counter = {}
    for s in diffs:
        if s not in unique_diffs:
            unique_diffs.append(s)
            counter[s] = 1
        else:
            counter[s] += 1
    return counter

spectra_shift_count = count_shifts(diffs)
random_pairs_shift_count = count_shifts(diffs2)
```

```python
spectra_shift_count_sorted = sorted(spectra_shift_count.items(), key=lambda x: x[1], reverse=True)
random_pairs_shift_count_sorted = sorted(random_pairs_shift_count.items(), key=lambda x: x[1], reverse=True)

print("list a)", spectra_shift_count_sorted)
print("=============")
print("list b)", random_pairs_shift_count_sorted)
```
